```python
#–ø—Ä–æ–≤–µ—Ä–∫–∞ –º–∞—Å–∏–≤–æ–≤ —Å –∫–æ—Ä—Ä–µ–∫—Ç–æ—Ä–∞–º–∏ 
inst.write("SYST:PRES")
def get_table(measure,ports):
    comand1=f"SENS1:CORR:COLL:{measure} {ports}"
    comand2=f"SENS1:CORR:COLL:DATA:{measure}? {ports}"
    print(comand1)
    inst.write(f"SENS1:CORR:COLL:{measure} {ports}")
    check_error()
    try:
        print(comand2)
        inst.query_ascii_values(f"SENS1:CORR:COLL:DATA:{measure}? {ports}")
        check_error()
    except:
        print('error')
        check_error()
inst.write(f"SENS:CORR:COLL:CKIT 2")
measure_types = ['ISOL','OPEN','LOAD','SHORT']
for measure in measure_types:
    if (measure=='ISOL'):
        get_table(measure,"1,2")
    else:
        get_table(measure,"1")
        
    
    
```
