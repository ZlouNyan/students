```python
def kalibr_N(con, Nchanal):
    con.timeout = 60000
    # –æ–ø–æ—Ä–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è —à–∏—Ä–∏–Ω—ã —Ñ–∏–ª—å—Ç—Ä–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫ –∏–∑–º–µ—Ä–µ–Ω–∏—è
    freq = 10000
    points = 1001
    con.write('SYST:CORR 0')
    con.write("SERVice:SWEep:FREQuency:FACTory")
    check_error()
    con.write("SENS:CORR:COLL:CKIT 2")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:ORD:SEL 4")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:STAN2:DEL 0")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:STAN2:LOSS 0")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:STAN2:C0 0")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:STAN2:C1 0")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:STAN2:C2 0")
    check_error()
    con.write("SENS:CORR:COLL:CKIT:STAN2:C3 0")
    check_error()

    con.write("DISP:WIND:TRAC:Y:PDIV 0.05")
    con.write("SENS:CORR:STAT ON")
    # time.sleep(2)
    i = 0
    while i < Nchanal:
        con.write("SENS:CORR:COLL:OPEN " + str(i + 1))
        con.write("SENS:CORR:COLL:METH:OPEN " + str(i + 1))
        con.write("SENS:CORR:COLL:SAVE")
        i = i + 1
    i = 0

    #con.write("SENSe:CORRection:COLLect:CKIT:RESet")
    con.write("SENSe:SWEep:TYPE LINear")
    return 0
```

```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from itertools import product,chain
def find_spur_v2(port,fstart,fstop,pointizm):
    global spur_table 
    print('### find spur')
    #print(port)
    #time.sleep(0.5)
    data = inst.query(f"CALC1:TRAC{port}:DATA:FDAT?")
    #time.sleep(1)
    check_error()
    #time.sleep(0.5)
    data = data.split(',')[::2] 
    frequency_data = inst.query("SENS:FREQ:DATA?").split(',')
    #print(f'data lenght = {len(data)} frequency_data lenght = {len(frequency_data)}')
    check_error()
    frequency_data = list(map(float,frequency_data))
    data = list(map(float, data))          
    def moving_average(x, w):        
        ext=np.pad(x, (int(w/2),0),'constant')        
        ext=np.pad(ext, (0,int(w/2)-1),'constant')        
        return np.convolve(ext, np.ones(w), 'valid') / w
    x= np.array(data)
    ampl=0.02
    #x[1000]=-0.014
    arr_averg=moving_average(x,20)
    #print(arr_averg.size)
    arr1=arr_averg+ampl
    arr2=arr_averg-ampl
    #print(arr_averg)
    #print(len(x),len(arr1))
    peaks, _ = find_peaks(x, height=arr1,prominence=ampl)
    negative_peaks, _ = find_peaks(-x, height=-arr2, prominence=ampl)
    if (len(peaks)!=0 or len(negative_peaks)!=0):
        print(f'{fstart} {fstop} {pointizm}')
#         print(peaks)
#         print(negative_peaks)
        spur_table.append((f'S{port}{port}',
                           fstart,
                           fstop,
                           pointizm,
                          [frequency_data[i] for i in chain(peaks,negative_peaks)]))
        plt.plot(x)
        plt.plot(arr1, color="gray")
        plt.plot(arr2, color="gray")
        plt.plot(peaks, x[peaks], "x")
        plt.plot(negative_peaks, x[negative_peaks], "x")
        plt.plot(np.zeros_like(x), "--", color="gray")
        plt.show()
        
    #peaks, _ = find_peaks(x, height=arr1)
    #negative_peaks, _ = find_peaks(-x, height=-arr2)
    #print(negative_peaks)
    #print(type(x),type(y))
    
```

```python
def table_generator(table):
    html_table='<table border="1">\
    <tr>\
        <td>–î–∏–∞–ø–∞–∑–æ–Ω</td>'
    td_background="bgcolor=\"#ffdd00\""    
    for key in admittance_table:
        start=admittance_table[key]['start']
        stop=admittance_table[key]['stop']
        
        html_table += f"<td colspan=3>{start} - {stop}</td>" 
    html_table += "</tr>"
        #######
    html_table += "<tr><td>–¢—Ä–∞—Å—Å—ã</td>"
    for i in range (int(range_count)):
        html_table += "<td>–ù–µ—Ä-—Ç—å,–¥–ë</td> \
                        <td>–ú–∞–∫—Å.–¥–ë</td> \
                        <td>–î–æ–ø—É—Å–∫,–¥–ë</td> "
    html_table += "</tr>"
    for i in table:
        html_table += "<tr>"
        html_table += f'<td>{i[0]}</td>'
        for j in i[1::]:
            for k in j:
                html_table += f'<td>{k[0]}</td>'  
                
                if k[1]>int(k[2]):                    
                    html_table +=f'<td {td_background}>{k[1]}</td>'
                    html_table +=f'<td {td_background}>{k[2]}</td>'
                    continue
                if (k[1]!=0):
                    html_table +=f'<td >{k[1]}</td>'
                    html_table +=f'<td >{k[2]}</td>'
                    continue
                else:
                    html_table +=f'<td>-</td>'
                    html_table +=f'<td>{k[2]}</td>'
                    
            
        
   
            
        html_table +='</tr>'
    html_table += "</table>"
    return html_table
```

```python
def html_header(data):
    html_table='<table border="1"><tr>'
    for key in data:
        html_table+=f'<td>{key}</td>'
    html_table+=f'</tr>'
    html_table+=f'<tr>'
    for key in data:
        html_table+=f'<td>{data[key]}</td>'
    html_table+=f'</tr>'    
    return html_table
```
